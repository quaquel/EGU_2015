# Analysis for objective function 3 : mean and undesirable deviation seperately

This notebook contains the analyses for the second objective function. This second objective functios is a satisfycing metric of robusntess. It is insipired by the signal to noise ratio, but rather than combining standard deviation and mean into a combined metric, it handles them as seperate objectives. The theoretical advantage of this is an increased insight into the tradeoff between mean performance and variation around this mean performance


$$
\begin{equation}
    \begin{aligned}
    & \text{Minimize} && F(l_{p,r}) = (\bar{y}_{damages}, f_{damages}, \bar{y}_{casualties}, f_{casualties}, \bar{y}_{costs}, f_{costs}) \\
    & && f_i (Q_1, \dots, Q_n) = \frac{1}{N}\sum_{n=1}^N(Q_{i,n}-\bar{y}_i)^2 [Q_n>\bar{y}_i] &&&i \in \{damages, casualities, costs\}    
    \end{aligned}
\end{equation}
$$

Where $Q_{i,n}$ is the score on the outcome indicitor $i$ in scenario $n$. Note that we are using iverson notation for the conditional sum. So, we are taking the sum of squares over all scenarios that performe worse than the mean performance, diveded by the total number of scenarios. 

```python
from expWorkbench import load_results, ema_logging

import scipy as sp

ema_logging.log_to_stderr(ema_logging.INFO)
results = load_results(r'./data/10 pathways 1000 MC experiments.tar.gz')

experiments, outcomes = results

outcomes["Flood damage (Milj. Euro)"] = np.sum(outcomes["Flood damage (Milj. Euro)"], axis=1)
outcomes["Number of casualties"] = np.sum(outcomes["Number of casualties"], axis=1)
```

```python
def obj_func3(outcomes):
    '''
    
    objective function that separates the mean and undesirable deviation
    from the mean as separate objectives. 
    
    '''
    
    # we want to minimize the median damage and the
    # dispersion of the damage
    # the lower the median, the lower score 1
    # the lower the dispersion, the lower the score
    outcome_1 = outcomes["Flood damage (Milj. Euro)"]
    score_1 = np.mean(outcome_1)

    variance_measure_1 = outcome_1[outcome_1>score_1]
    variance_measure_1 = np.power(score_1-variance_measure_1, 2)
    variance_measure_1 = np.sum(variance_measure_1)/outcome_1.shape[0]
    
    if score_1 >= 50000:
        # the constraints are hard, that is, crossing one of them
        # produces infinity on all criteria
        return (np.inf,) * 6
    
    # we want to minimize the median casualties and the
    # dispersion of the casualties
    # the lower the median, the lower score 1
    # the lower the dispersion, the lower the score
    outcome_2 = outcomes["Number of casualties"]
    score_2 = np.mean(outcome_2)

    variance_measure_2 = outcome_2[outcome_2>score_2]
    variance_measure_2 = np.power(score_2-variance_measure_2, 2)
    variance_measure_2 = np.sum(variance_measure_2)/outcome_2.shape[0]
    
    if score_2 >= 1000:
        return (np.inf,) * 6
       
    # we want to minimize the median costs and the
    # dispersion of the costs
    # the lower the median, the lower score 1
    # the lower the dispersion, the lower the score
    outcome_3 = outcomes["Costs"]
    score_3 = np.mean(outcome_3)

    variance_measure_3 = outcome_3[outcome_3>score_3]
    variance_measure_3 = np.power(score_3-variance_measure_3, 2)
    variance_measure_3 = np.sum(variance_measure_3)/outcome_3.shape[0]

  
    return score_1, variance_measure_1,\
           score_2, variance_measure_2,\
           score_3, variance_measure_3
```

```python
experiments, outcomes = results
pathways = set(experiments['policy'])

flood_damage_mean = {}
flood_damage_std = {}
casualties_mean = {}
casualties_std = {}
costs_mean = {}
costs_std = {}

for pathway in pathways:
    logical = experiments['policy']==pathway
    temp_outcomes = {key: value[logical] for key, value in outcomes.iteritems()}
    nr_experiments = np.sum(logical)
    
    temp_flood_damage_mean = []
    temp_flood_damage_std = []

    temp_casualties_mean = []
    temp_casualties_std = []
    
    temp_costs_mean = []
    temp_costs_std = []
    temps = [temp_flood_damage_mean, temp_flood_damage_std, 
             temp_casualties_mean, temp_casualties_std,
             temp_costs_mean, temp_costs_std]
    
    for  i in range(1, nr_experiments):
        temp_i_outcomes = {key: value[0:i] for key, value in temp_outcomes.iteritems()}

        scores= obj_func3(temp_i_outcomes)

        for i, score in enumerate(scores):
            temps[i].append(score)
            
    flood_damage_mean[pathway] = temp_flood_damage_mean
    flood_damage_std[pathway] = temp_flood_damage_std
    casualties_mean[pathway] = temp_casualties_mean
    casualties_std[pathway] = temp_casualties_std
    costs_mean[pathway] = temp_costs_mean
    costs_std[pathway] = temp_costs_std
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker  

def make_parallel_plot(nr_columns, labels, maxima, minima):

    fig = plt.figure()
    axes = []
    
    # we need one axes less than the shape
    for i in range(1, nr_columns):
        ax = fig.add_subplot(1,nr_columns-1,i,  ylim=(-0.1,1.1))
        axes.append(ax)
        ax.set_xlim([i,i+1])
        ax.xaxis.set_major_locator(ticker.FixedLocator([i]))
        ax.xaxis.set_ticklabels([labels[i-1]], rotation=45)
        ax.xaxis.set_tick_params(bottom=False, top=False)
        
        #let's put our own tick labels
        ax.yaxis.set_ticks([])
        ax.text(i, 1.01, "{:.2f}".format(maxima[i-1]), va="bottom", ha="center")
        ax.text(i, -0.01,"{:.2f}".format(minima[i-1]), va="top", ha="center")
        
        ax.spines['left'].set_bounds(0, 1)
        ax.spines['right'].set_bounds(0, 1)
        ax.spines['top'].set_visible(False)
        ax.spines['bottom'].set_visible(False)

    
    # for the last axis, we need 2 ticks (also for the right hand side
    ax.spines['right'].set_bounds(0, 1)
    ax.xaxis.set_major_locator(ticker.FixedLocator([i, i+1]))
    ax.xaxis.set_ticklabels(labels[i-1:i+1])
    ax.text(i+1, 1.01, "{:.2f}".format(maxima[i]), va="bottom", ha="center")
    ax.text(i+1, -0.01,"{:.2f}".format(minima[i]), va="top", ha="center")
    
    # add the tick labels to the rightmost spine
    for tick in ax.yaxis.get_major_ticks():
        tick.label2On=True
    
    # stack the subplots together
    plt.subplots_adjust(wspace=0)
    
    return axes


def parallel_pareto_front(stats_callback, labels, with_tried=True):
    
    def normalize(data, data_to_norm):
        minima = np.min(data, axis=0)
        maxima = np.max(data, axis=0)
        d = maxima - minima
        
        d[d==0] = 1
        
        norm_data = data_to_norm/d - minima/d
        
        minima = np.asarray([float(entry) for entry in minima])
        maxima = np.asarray([float(entry) for entry in maxima])
        
        return norm_data, minima, maxima
    

    hof = stats_callback.archive
    tried = stats_callback.stats
    if type(tried)==pd.DataFrame:
        tried = tried.as_matrix()
    
    data_hof = np.zeros((len(hof), len(hof[0].fitness.values)))
    
    for i, entry in enumerate(hof):
        data_hof[i,:] = entry.fitness.values    
    
    if with_tried:
        hof_norm, minima, maxima = normalize(tried, data_hof)
        tried_norm, minima, maxima = normalize(tried, tried)
    else:
        hof_norm, minima, maxima = normalize(data_hof, data_hof)
    
    axes = make_parallel_plot(hof_norm.shape[1], labels, maxima, minima)
    
    if with_tried:
        # visualize tried solution
        for i,j in zip(range(tried_norm.shape[1]-1),range(1,tried_norm.shape[1])):
            ax = axes[i]
            y = tried_norm[:, i:j+1]
            x = np.tile([i+1,j+1], (tried_norm.shape[0], 1))
            ax.plot(x.T, y.T, c='grey', alpha=0.2, lw=1)
    
    # visualize hof
    for i,j in zip(range(hof_norm.shape[1]-1),range(1,hof_norm.shape[1])):
        ax = axes[i]
        y = hof_norm[:, i:j+1]
        x = np.tile([i+1,j+1], (hof_norm.shape[0], 1))
        ax.plot(x.T, y.T, lw=2)

        
obj_labels = ['$\hat{x}_{damages}$', '$s_{damages}$',
              '$\hat{x}_{casualties}$', '$s_{casualties}$',
              '$\hat{x}_{costs}$', '$s_{costs}$',]
parallel_pareto_front(stats, obj_labels)
parallel_pareto_front(stats, obj_labels, with_tried=False)
```
